# Unsupervised Learning of Quantum System's Phases from Quantum Simulation Data

In this notebook, we demonstrate how unsupervised learning (UL) can be used to classify the phases of a quantum system.

The three datasets in use here all contain a number of classical shadows, which can be thought of as approximate descriptors of quantum states. The first two datasets hold ground state classical shadows. For the Z-dataset, the random measurements on qubits are all performed in the Z-direction, while for the CS-dataset, random Pauli measurements ($\sigma_x$, $\sigma_y$, and $\sigma_z$) were performed on the qubits. The third dataset contains time averaged classical shadows (TACS), which are classical shadows of the Von Neumann ensemble obtained from dynamics simulation. Since quantum dynamics is naturally simulated on a quantum computer, we want to see if the ML method can learn about the phase diagram from dynamics data.

The system being studied here is the Transverse field Ising Model (TFIM) (100 spins in the ground state study, 20 spins in the dynamics study). First two datasets were obtained by the density matrix renormalization group (DMRG) quantum simulation, while the third was obtained by a simulation of the system's quantum dynamics.  



The UL method being applied here is called diffusion maps, it's a nonlinear kernel-based dimensionality reduction method where the manifold on which the datapoints are located is learned. We can assign a random walk transition probability function based on the kernel function and construct a transition matrix $P$. The number of nontrivial large eigenvalues of the $P$-matrix is the number of effective dimensions, and projecting the datapoints on it's eigenvectors results in an effective lower dimensional representation of the dataset.

For a detailed explanation, I recommend our published article: https://journals.aps.org/prb/abstract/10.1103/PhysRevB.108.235141)

```python
import os
import re
from math import sqrt
import numpy as np
import pandas as pd
from numba import njit
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.linalg import eigh
from sklearn.manifold import MDS
```

```python
# loading and encoding data
def extract_num(string):
    """
    Extracts numbers from a given string.
    """
    result = [float(num) for num in re.findall(r"[-+]?\d*\.\d+|\d+", string)]
    return result[0] if len(result) == 1 else result

def cs_encoding(x):
    """
    Encodes qubit states in a string to a list of integers.
    """
    x = ''.join(filter(str.isalnum, x.replace('"', '').replace('=', '').replace('>', '')))
    encoding_map = {'Z1': 0, 'Z0': 1, 'X1': 2, 'X0': 3, 'Y1': 4, 'Y0': 5}
    return [encoding_map.get(x[2*i:2*i+2], -1) for i in range(len(x)//2)]

def load_z_data(datadir_z_groundstate):
    """Load Z ground state data from specified directory."""
    files_z = os.listdir(datadir_z_groundstate)
    df_z = [pd.read_csv(datadir_z_groundstate + f, sep='\t', names=['Sz']) for f in files_z]
    for i, df in enumerate(df_z):
        df['hx'] = float(files_z[i][3:-4])
    return df_z

def load_cs_data(datadir_cs_groundstate):
    """Load CS ground state data from specified directory."""
    files_cs = os.listdir(datadir_cs_groundstate)
    df_cs = [pd.read_csv(datadir_cs_groundstate + f, sep='\t', names=['spin measurements']) for f in files_cs]
    for i, df in enumerate(df_cs):
        df['hx'] = float(files_cs[i][3:-4])
        df['spin measurements'] = df['spin measurements'].apply(cs_encoding)
    return df_cs

def load_dnmks_data(datadir):
    """
    Loads TACS data from specified directory, applying filtering and custom encoding.
    """
    files = [f for f in os.listdir(datadir) if 'ti' not in f]
    df_lst = []
    for f in files:
        hx = extract_num(f)
        df = pd.read_csv(os.path.join(datadir, f), sep='\t', names=['spin_measurements'])
        df['spin_measurements'] = df['spin_measurements'].apply(cs_encoding)
        df['hx'] = hx
        df_lst.append(df)
    return df_lst

def process_z_data(data_frame, no_of_snapshots):
    """Process Z data frames to extract state arrays and hx values."""
    state_list = []
    states = np.zeros((len(data_frame), no_of_snapshots, 100))
    hx_values = np.zeros(len(data_frame) * no_of_snapshots)
    ind = np.random.choice(np.arange(data_frame[0].shape[0]), size=no_of_snapshots, replace=False)
    for i, df in enumerate(data_frame):
        state_list.append(df.loc[ind]['Sz'].apply(lambda x: np.array(list(map(int, list(x))))).to_numpy())
        hx_values[i * no_of_snapshots: (i + 1) * no_of_snapshots] = df.loc[ind]['hx'].apply(lambda x: float(x)).to_numpy()
    for i, states_batch in enumerate(state_list):
        for j, state in enumerate(states_batch):
            states[i][j][:] = state
    return states.astype("int"), hx_values

def process_cs_data(data_frame, no_of_snapshots):
    """Process CS data frames to extract state arrays and hx values."""
    state_list = []
    states = np.zeros((len(data_frame), no_of_snapshots, 100))
    hx_values = np.zeros(len(data_frame) * no_of_snapshots)
    ind = np.random.choice(np.arange(data_frame[0].shape[0]), size=no_of_snapshots, replace=False)
    for i, df in enumerate(data_frame):
        state_list.append(df.loc[ind]['spin measurements'].apply(lambda x: np.array(x)).to_numpy())
        hx_values[i * no_of_snapshots: (i + 1) * no_of_snapshots] = df.loc[ind]['hx'].apply(lambda x: float(x)).to_numpy()
    for i, states_batch in enumerate(state_list):
        for j, state in enumerate(states_batch):
            states[i][j][:] = state
    return states.astype("int"), hx_values

def process_dnmks_data(data_frame):
    """
    Processes a list of DataFrames, extracting and sorting by 'hx' values.
    """
    state_list = []
    h = len(data_frame)
    T = len(data_frame[0])
    N = len(data_frame[0]['spin_measurements'].loc[0])
    states = np.zeros((h ,T, N))
    hx_values = []
    for i in range(len(data_frame)):
      state_list.append(data_frame[i]['spin_measurements'].apply(lambda x: np.array(x)).to_numpy())
      hx_values.append(data_frame[i]['hx'].loc[0])
    for i in range(len(state_list)):
        for j in range(len(state_list[0])):
            states[i][j][0:N] = state_list[i][j][0:N]
    idx = np.argsort(hx_values)
    states = states[idx]
    hx_values = np.array(hx_values)[idx]
    return states.astype("int"), hx_values
```

Building the shadow kernel matrix from the shadow kernel function in: https://www.nature.com/articles/s42254-021-00411-5

```python
#construct the shadow kernel matrix for each dataset from the shadow kernel function
def sigma(vec):
    return 3*np.outer(vec, np.conjugate(vec)) - np.identity(len(vec))

def trace_sigma(vec1,vec2):
    return np.trace(np.matmul(sigma(vec1), sigma(vec2)))

@njit
def trace_mul(snap1,snap2,trace_matrix):
    sum = 0
    for i in range(len(snap1)):
        sum = sum + trace_matrix[snap1[i],snap2[i]]
    return sum

@njit
def shadow_kernel(rho_1,rho_2,trace_matrix,tau,gamma):
    c = 0
    n = len(rho_1[0])
    T = len(rho_1)
    for i in range(len(rho_1)):
        for j in range(len(rho_2)):
            c = c + np.exp((gamma/n)*trace_mul(rho_1[i],rho_2[j],trace_matrix))
    return np.exp(c*(tau/T**2))

@njit
def kernel_matrix(state_set, trace_matrix, tau, gamma):
  K = np.zeros((len(state_set),len(state_set)))
  for i in range(len(state_set)):
    for j in range(len(state_set)):
      K[i,j] = shadow_kernel(state_set[i],state_set[j], trace_matrix, tau, gamma)
  return K
```

```python
# P-matrix eigenvalues and eigenvectors give the number of effective dimensions and diffusion coordinates
def find_diffusion_matrix(Kernel, alpha=0.15):
    """Function to find the diffusion matrix P

        >Parameters:
        alpha - to be used for gaussian kernel function
        X - feature matrix as numpy array

        >Returns:
        P_prime, P, Di, K, D_left
    """

    r = np.sum(Kernel, axis=0)
    Di = np.diag(1/r)
    P = np.matmul(Di, Kernel)

    D_right = np.diag((r)**0.5)
    D_left = np.diag((r)**-0.5)
    P_prime = np.matmul(D_right, np.matmul(P,D_left))

    return P_prime, P, Di, Kernel, D_left

def find_diffusion_map(P_prime, D_left):
    """Function to find the diffusion coordinates in the diffusion space

        >Parameters:
        P_prime - Symmetrized version of Diffusion Matrix P
        D_left - D^{-1/2} matrix

        >Returns:
        Diffusion_map as np.array object
    """
    # n_eign = n_eign

    eigenValues, eigenVectors = eigh(P_prime)
    idx = eigenValues.argsort()[::-1]
    eigenValues = eigenValues[idx]
    eigenVectors = eigenVectors[:,idx]
    diffusion_coordinates = np.matmul(D_left, eigenVectors)

    return eigenValues, diffusion_coordinates
```

```python
# Example usage:
if __name__ == "__main__":
    datadir_dnmks = '/path/to/dynamics/dataset/'  # Update this to your dataset path
    datadir_z_groundstate = '/path/to/z-groundstate/dataset/'
    datadir_cs_groundstate = '/path/to/cs-groundstate/dataset/'

    df_list_z = load_z_data(datadir_z_groundstate)
    df_list_cs = load_cs_data(datadir_cs_groundstate)

    states_z, hx_values_z = process_z_data(df_list_z, no_of_snapshots=1000)
    states_cs, hx_values_cs = process_cs_data(df_list_cs, no_of_snapshots=1000)

    df_list_dnmks = load_dnmks_data(datadir_dnmks)
    tacs_states_dnmks, hx_vals_dnmks = process_dnmks_data(df_list_dnmks)
    print("Data processed.")
```

```python
if __name__ == "__main__":
    # pauli matrix eigenvectors
    z_up = np.array([1,0])
    z_down = np.array([0,1])
    x_up = (1/sqrt(2))*np.array([1,1])
    x_down = (1/sqrt(2))*np.array([1,-1])
    y_up = (1/sqrt(2))*np.array([1, 1j])
    y_down = (1/sqrt(2))*np.array([1,-1j])
    cs_eigvecs = [z_up,z_down,x_up,x_down,y_up,y_down]
    z_eigvecs = [z_down, z_up]
    tr_mat_cs = np.zeros((len(cs_eigvecs),len(cs_eigvecs)))
    for i in range(len(cs_eigvecs)):
        for j in range(len(cs_eigvecs)):
            tr_mat_cs[i,j] = trace_sigma(cs_eigvecs[i], cs_eigvecs[j])
    tr_mat_z = np.zeros((len(z_eigvecs),len(z_eigvecs)))
    for i in range(len(z_eigvecs)):
        for j in range(len(z_eigvecs)):
            tr_mat_z[i,j] = trace_sigma(z_eigvecs[i], z_eigvecs[j])
    kernel_matrix(states_cs[0:5],tr_mat_cs,1,1)
    K_z = kernel_matrix(states_z,tr_mat_z,0.8,0.8)  #z-dataset kernel matrix
    K_cs = kernel_matrix(states_cs,tr_mat_cs,1,1)   # cs-dataset kernel matrix
    K_tacs = kernel_matrix(tacs_states_dnmks,tr_mat_cs,1,1)  #TACS dataset kernel matrix
```

```python
#ground state z-data results
if __name__ == "__main__":
    P_prime, P, Di, K, D_left = find_diffusion_matrix(K_z, alpha=0.15)
    eigenValues, diff_coords = find_diffusion_map(P_prime, D_left)

    fig, axs = plt.subplots(1, 2, figsize=(14, 6)) # You can adjust figsize to fit your needs

    # First subplot for eigenvalues
    axs[0].scatter(range(0, len(eigenValues))[1:11], eigenValues[1:11], c='b', s=150)
    axs[0].set_xlabel(r"$k$", fontsize=25)
    axs[0].set_ylabel(r"$\lambda_{k}$", fontsize=25)
    axs[0].tick_params(axis='both', which='major', labelsize=15)
    axs[0].set_xticks(ticks=[1, 6, 10])

    # Second subplot for MDS scatter plot
    X = diff_coords[:, 1:6]
    embedding = MDS(n_components=2)
    X_transformed = embedding.fit_transform(X)
    mds1 = X_transformed[:, 0]
    mds2 = X_transformed[:, 1]
    group = hx_values_z[::1000]
    greater = np.where(group > 1)[0]
    smaller = np.where(group < 1)[0]

    axs[1].scatter(mds1[greater], mds2[greater], label='$h_{x} > 1.0$', s=120, c='tab:orange', marker='s', alpha=0.5)
    axs[1].scatter(mds1[smaller], mds2[smaller], label='$h_{x} < 1.0$', s=120, c='b', marker='v', alpha=0.5)
    axs[1].legend(loc='center left', fontsize=20, edgecolor='0.7')
    axs[1].set_xlabel('ax1', fontsize=25)
    axs[1].set_ylabel('ax2', fontsize=25)
    axs[1].tick_params(axis='both', which='major', labelsize=15)
    axs[1].set_xticks(ticks=[0.0])
    axs[1].set_xticklabels(labels=[r'$0.0$'])
    axs[1].set_yticks(ticks=[0.0])
    axs[1].set_yticklabels(labels=[r'$0.0$'])

    plt.show()
```

The eigenvalue plot above shows five nontrivial large eigenvalues, hence we choose the first five diffusion coordinates to project our data onto. The figure on the right shows a two-dimensional embedding of this five dimensional data (done using multidimensional scaling (MDS).) We see that the datapoints cluster nicely
with respect to the phases. The blue datapoints represent the ferromagnetic phase and the orange ones represent the paramagnetic phase. This success isn't surprising, since the z-measurements directly relate to the order parameter ($M_z$, the z-magnetization) and it's the "best angle" to look at the quantum state from in this case.  

```python
#ground state CS-data results
if __name__ == "__main__":
    P_prime, P, Di, K, D_left = find_diffusion_matrix(K_cs, alpha=0.15)
    eigenValues, diff_coords = find_diffusion_map(P_prime, D_left)

    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    # First subplot for eigenvalues
    axs[0].scatter(range(0, len(eigenValues))[1:8], eigenValues[1:8], c='b', s=150)
    axs[0].set_xlabel(r"$k$", fontsize=25)
    axs[0].set_ylabel(r"$\lambda_{k}$", fontsize=25)
    axs[0].tick_params(axis='both', which='major', labelsize=15)
    axs[0].set_xticks(ticks=[1, 4, 7])

    # Second subplot for diffusion map scatter plot
    dc1 = diff_coords[:,1]
    dc2 = diff_coords[:,2]
    group = hx_values_cs[::1000]
    greater = np.where(group > 1)[0]
    smaller = np.where(group < 1)[0]

    axs[1].scatter(dc2[greater], dc1[greater], label='$h_{x} > 1.0$', s=120, c='tab:orange', marker='s', alpha=0.5)
    axs[1].scatter(dc2[smaller], dc1[smaller], label='$h_{x} < 1.0$', s=120, c='b', marker='v', alpha=0.5)
    axs[1].legend(loc='center left', fontsize=20, edgecolor='0.7')
    axs[1].set_xlabel('dc2', fontsize=25)
    axs[1].set_ylabel('dc1', fontsize=25)
    axs[1].tick_params(axis='both', which='major', labelsize=15)
    axs[1].set_xticks(ticks=[0.0])
    axs[1].set_xticklabels(labels=[r'$0.0$'])
    axs[1].set_yticks(ticks=[0.0])
    axs[1].set_yticklabels(labels=[r'$0.0$'])

    plt.show()
```

The above plot shows, in the case of the randomized CS-dataset, there are only two dominant eigenvalues. So we project the datapoints onto the first two diffusion coordinates, and the resulting image (right) looks a lot like symmetry breaking phase transition. Which is not a coincidence, the diffusion coordinates dc1 and dc2 are correlated with system parameters $M_z$ (z-magnetization) and $h_x$ (transverse field). It looks like diffusion map can classify phases by "learning" the relevant observables and parameters, through having access to a large portion of the phase diagram.  

```python
# TACS data diffusion map results
if __name__ == "__main__":
    P_prime, P, Di, Kernel, D_left = find_diffusion_matrix(K_tacs, alpha=0.15)
    eigenValues, diff_coords = find_diffusion_map(P_prime, D_left)

    fig, axs = plt.subplots(1, 2, figsize=(13, 6))

    # Eigenvalues scatter plot
    axs[0].scatter(range(1, 8), eigenValues[1:8], s=60, c='b')
    axs[0].set_xlabel(r"$k$", fontsize=20)
    axs[0].set_ylabel(r"$\lambda_{k}$", fontsize=20)
    axs[0].set_xticks([1, 4, 7])
    axs[0].set_yticks([0, 0.017, 0.035])
    axs[0].set_yticklabels([r'$0.0$', r'$1.7$', r'$3.5$'], fontsize=10)
    axs[0].set_ylim(0, 0.043)
    axs[0].set_xlim(0.25, 7.75)
    axs[0].text(0.5, 0.045, r'$\times 10^{-2}$', fontsize=10)

    # Scatter plot of dc1 vs dc2
    group = hx_vals_dnmks
    greater = np.where(group > 1)[0]
    smaller = np.where(group < 1)[0]
    z1 = np.log(hx_vals_dnmks[greater])
    z2 = np.log(hx_vals_dnmks[smaller])
    min_, max_ = min(z1.min(), z2.min()), max(z1.max(), z2.max())

    dc1, dc2 = diff_coords[:, 1], diff_coords[:, 2]
    scat1 = axs[1].scatter(dc1[greater], dc2[greater], c=z1, cmap='jet', s=60, marker='v', alpha=0.5)
    scat2 = axs[1].scatter(dc1[smaller], dc2[smaller], c=z2, cmap='jet', s=60, marker='v', alpha=0.5)

    scat1.set_clim(min_, max_)
    scat2.set_clim(min_, max_)

    cb = fig.colorbar(scat1, ax=axs[1], fraction=0.046, pad=0.04)
    cb.set_label(label=r'$\log(h_{x})$', size=13)
    cb.ax.tick_params(labelsize=10)

    axs[1].set_xlabel('dc1', fontsize=20)
    axs[1].set_ylabel('dc2', fontsize=20)
    axs[1].set_xticks([0.0])
    axs[1].set_xticklabels([r'$0.0$'], fontsize=10)
    axs[1].set_yticks([0.0])
    axs[1].set_yticklabels([r'$0.0$'], fontsize=10)
    axs[1].text(0.0009, 0.0012, 'quantum\ncritical\nregion', fontsize=9, ha='left')

    plt.tight_layout()
    plt.show()
```

Performing diffusion maps on the TACS dataset reveals the existence of a quantum critical region. In this case, there's no clear clustering because the initial states in the quantum dynamics were chosed from a region of the phase diagram where there's no clear phase boundary between the paramagnetic and ferromagnetic phase. We later found through Bayesian inference calculation that dc1 here is related to the renyi entropy of the system.
